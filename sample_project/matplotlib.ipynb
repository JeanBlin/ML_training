# Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # Width, height

# 3. Plot the data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & Show (you save the whole figure)
fig.savefig('images/sample-plot.png')
```

# Which plotting method should we use

* When plotting quickly: use pyplot method
* When plotting more advanced, use OO 

```python
hd.head()
```

```python
over_50 = hd[hd['age'] > 50]
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind='scatter', 
             x='age',
             y='chol',
             c='target')
```

```python
# OO method (ish)
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

```python
# OO method (True)
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize
ax.set(title='Heart Disease and cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol')

# Add legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add horizontal line for medium value
ax.axhline(y=over_50['chol'].mean(),
          linestyle='--')
```
